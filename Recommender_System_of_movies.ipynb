```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
rating = pd.DataFrame(df.groupby('title')['rating'].mean())
```

```python
rating.head()
```

```python
rating['number of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
```

```python
rating.head()
```

```python
moviemat = df.pivot_table(index='User_id',columns='title',values='rating')
```

```python
moviemat.head()
```

```python
rating.sort_values('number of ratings',ascending=False).head(10)
```

```python
starwarrating = moviemat['Star Wars (1977)']
```

```python
liarliar = moviemat['Liar Liar (1997)']
```

```python
starwarrating.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwarrating)
```

```python
similar_to_starwars
```

```python
similar_to_liarliar = moviemat.corrwith(liarliar)
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['correlation'])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlation',ascending=False).head(10)
```

```python
corr_starwars = corr_starwars.join(rating['number of ratings'])
```

```python
corr_starwars.head(10)
```

```python
corr_starwars[corr_starwars['number of ratings']>100].sort_values('correlation',ascending=False)
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['correlation'])
```

```python
corr_liarliar.dropna(inplace=True)
```

```python
corr_liarliar.head()
```

```python
corr_liarliar = corr_liarliar.join(rating['number of ratings'])
```

```python
corr_liarliar.head()
```

```python
corr_liarliar[corr_liarliar['number of ratings']>100].sort_values('correlation',ascending=False).head(10)
```
